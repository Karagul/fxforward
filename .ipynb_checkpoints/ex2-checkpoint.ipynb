```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from fxforward import FXForward, ee, pfe
```

```python
plt.style.use("ggplot")
```

```python
sim_data_eur = pd.read_csv("sim_eurgbpfx.csv")
sim_data_eur["Date"] = pd.to_datetime(sim_data_eur["Date"])
```

```python
sim_data_eur = sim_data_eur.set_index(["Date"]).drop(columns="Unnamed: 0")
```

```python
f = FXForward("GBP","EUR",10000,1.368*10000)
```

```python
sim_data_gbp = pd.DataFrame(index=range(len(sim_data_eur)), columns=sim_data_eur.columns)
sim_data_gbp.iloc[:] =1
```

```python
fxres = f.calc_simulations(sim_data_gbp, sim_data_eur)
```

```python
ee(fxres).plot(label="EE")
pfe(fxres, 0.5).plot(label="PFE 0.5")
pfe(fxres, 0.90).plot(label="PFE 0.90")
_ = plt.legend()
```

```python
CURRENT_RATE = 1.352491
def mk_random_trades(n_trades=10):
    trades = []
    for i in range(n_trades):
        pay = (np.random.random()+1)*10000
        f = FXForward("GBP", "EUR", pay, CURRENT_RATE*pay)
        trades.append(f)
    return trades
portfolio = mk_random_trades()
```

```python
values = [t.calc_simulations(sim_data_gbp, sim_data_eur) for t in portfolio]
```

```python
EE = pd.Series(index=range(261))
PFE5 = pd.Series(index=range(261))
PFE90 = pd.Series(index=range(261))
EE.iloc[:]=0
PFE5.iloc[:]=0
PFE90.iloc[:]=0
for val in values:
    EE += ee(val)
#     ee(val).plot()
    PFE5 += pfe(val, 0.5)    
    PFE90 += pfe(val, 0.9)    
```

```python
EE.plot()
PFE5.plot()
PFE90.plot()
```
