```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("eurofxref-hist.csv")
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
filter_2015 = df["Date"] <= pd.to_datetime("31-12-2015")
filter_2013 = df["Date"] >= pd.to_datetime("1-1-2013")
filter_both = filter_2013 & filter_2015
```

```python
df_pre = df[["Date", "JPY", "USD","GBP", "CHF", "AUD", "CAD"]][filter_both]
df_pre.head()
```

```python
df_gbp_base = pd.DataFrame(columns=["Date", "GBP", "EUR", "JPY", "USD", "CHF", "AUD", "CAD"])
df_gbp_base["Date"] = df_pre["Date"]
```

```python
df_gbp_base["GBP"] = 1
df_gbp_base["EUR"] = 1 / df_pre["GBP"]
df_gbp_base["JPY"] = df_pre["JPY"] / df_pre["GBP"]
df_gbp_base["USD"] = df_pre["USD"] / df_pre["GBP"]
df_gbp_base["CHF"] = df_pre["CHF"] / df_pre["GBP"]
df_gbp_base["JPY"] = df_pre["JPY"] / df_pre["GBP"]
df_gbp_base["AUD"] = df_pre["AUD"] / df_pre["GBP"]
df_gbp_base["CAD"] = df_pre["CAD"] / df_pre["GBP"]

df_gbp_base.head()
```

```python
del df 
del df_pre
```

```python
df_gbp_eur = df_gbp_base.sort_values(by=["Date"])
```

```python
for col in df_gbp_eur.columns[1:]:
    df_gbp_eur[col+"_next"] = df_gbp_eur[col].shift(-1)
```

```python
for col in df_gbp_eur.columns[1:7]:
    df_gbp_eur[col+"_logret"] = np.log(df_gbp_eur[col+"_next"] / df_gbp_eur[col])
df_gbp_eur
```

```python
# 1: indexing cuz GBP is not needed
df_logrets = df_gbp_eur[list(filter(lambda c : c[-6:] == "logret", df_gbp_eur.columns))[1:]]
df_logrets.corr()
```

```python
df_eur_only = df_gbp_eur[["Date","EUR", "EUR_logret"]]
```

```python
df_eur_only.tail()
```

```python
def generate_eur():
    log_std = df_eur_only["EUR_logret"].std()
    # 1000 sampling from gaussian
    gauss_sample = np.random.normal(0,log_std,size=[1000,261])
    last_fx = df_eur_only["EUR"].iloc[-1]
    sim_fx = np.empty([1000, 261])

    for i_sim in range(1000):
        curr_fx = last_fx
        for i_day in range(261):
            a = np.exp(gauss_sample[i_sim,i_day])
            sim_fx[i_sim, i_day] = curr_fx*a
            curr_fx = sim_fx[i_sim, i_day]
    def weekday(date):
        wday = date.isoweekday()
        return wday<6

    df_sim = pd.DataFrame(columns=["Date"])
    df_sim["Date"] = pd.date_range("1-1-2016", "31-12-2016")
    df_sim = df_sim[df_sim["Date"].apply(weekday)]
    for i in range(1000):
        df_sim["Sim"+str(i)] = sim_fx[i]
    return df_sim
# generate_eur()["Sim0"].plot()
```

# Cholesky decomposition

```python
df_logrets = df_gbp_eur[list(filter(lambda c : c[-6:] == "logret", df_gbp_eur.columns))[1:]]
corr_mat = df_logrets.corr()
```

```python
L = np.linalg.cholesky(corr_mat)
```
